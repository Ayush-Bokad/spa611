## Question 1

**Verify that the mean of power is centered around the variance of the voltage**

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

# Number of samples
N = 100000

# Generate zero-mean Gaussian voltage (RMS = 2)
sigma = 2
v = np.random.normal(0, sigma, N)

# Compute mean power and variance
mean_power = np.mean(v**2)
variance = np.var(v)
mean_voltage = np.mean(v)

print("Mean of voltage:", mean_voltage)
print("Mean Power:", mean_power)
print("Variance:", variance)
```

We produced a zero-mean Gaussian voltage signal.

For such a signal, the mean of the squared voltage (power) should be equal to the variance.

And from our results, the mean voltage is approximately zero, and the mean power is almost equal to the variance.
This confirms that for a zero-mean random signal, mean power equals variance.

## Question 2
**Consider cases of different RMS and plot their histograms**

```python
N = 100000
sigmas = [1, 2, 4]

plt.figure()

for sigma in sigmas:
    v = np.random.normal(0, sigma, N)
    plt.hist(v, bins=100, density=True, alpha=0.5, label=f"RMS={sigma}")

plt.legend()
plt.title("Voltage Histograms for Different RMS Values")
plt.xlabel("Voltage")
plt.ylabel("Probability Density")
plt.show()
```

We produced Gaussian voltage signals with varying RMS values.

When the RMS value is increased:

- The distribution becomes wider.

- The peak becomes lower in height.

- The overall Gaussian shape remains the same.

The shape of the Gaussian distribution does not change.

## Question 3
**How will the histogram change if I add a slowly varying sine wave?**

```python
t = np.linspace(0, 10, N)

noise = np.random.normal(0, 1, N)
sine = 2 * np.sin(0.5 * t)

v_total = noise + sine

plt.figure()
plt.hist(v_total, bins=100, density=True)
plt.title("Histogram after Adding Slowly Varying Sine Wave")
plt.xlabel("Voltage")
plt.ylabel("Density")
plt.show()
```

When the slowly varying sine wave is added to the noise, the signal is no longer random. The histogram will be wider and slightly irregular compared to the Gaussian noise. This is because we are adding random noise to a periodic signal.

## Question 4
**How are the variances of the ensembles distributed?**

```python
num_ensembles = 500
ensemble_size = 1000

variances = []

for _ in range(num_ensembles):
    v = np.random.normal(0, 2, ensemble_size)
    variances.append(np.var(v))

plt.figure()
plt.hist(variances, bins=50, density=True)
plt.title("Distribution of Sample Variances")
plt.xlabel("Variance")
plt.ylabel("Density")
plt.show()

print("Mean of estimated variances:", np.mean(variances))
```

We calculated the variance for many independent ensembles.

Each calculated variance is slightly different because it is estimated from finite samples. The histogram shows that these variances are centered around the true variance, with some spread due to statistical fluctuations. If we increase the number of samples per ensemble, the spread would decrease.
